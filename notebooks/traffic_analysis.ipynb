```python
import pandas as pd
df = pd.read_csv(r"E:\urban-traffic-analysis\data\traffic_data.csv.txt")
```

```python
import matplotlib.pyplot as plt
```

```python
df.shape
```

```python
df.columns
```

```python
df["hour"] = df["hour"]
df["vehicle_count"] = df["vehicle_count"]

hourly_traffic = df.groupby("hour")["vehicle_count"].mean()
```

```python
df.head(11)
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    hourly_traffic.index,
    hourly_traffic.values,
    marker="o",
    linestyle="-",
    linewidth=2,
    alpha=0.8
)

plt.title("Average Hourly Traffic Volume", fontsize=14, fontweight="bold")
plt.xlabel("Hour of Day", fontsize=11)
plt.ylabel("Average Vehicle Count", fontsize=11)

plt.xticks(hourly_traffic.index)
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.savefig("E:\\urban-traffic-analysis\plots\hourly_traffic_volume.png", dpi=300)
plt.show()
```
